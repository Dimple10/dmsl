```python
import sampler as sm
import survey
from importlib import reload 
import mass_function as mf
import mass_profile as mp
import numpy as np
import matplotlib.pyplot as plt
import astropy.units as u
import scipy
%load_ext autoreload
%autoreload 2
%matplotlib inline
```

```python
r_high = survey.Roman(maxdlens=5*u.kpc, alphasigma=0.001*u.uas/u.yr**2)
r_norm = survey.Roman(maxdlens=1*u.kpc, alphasigma=0.1*u.uas/u.yr**2)
```

```python
%%time
cdm = mf.CDM_Test(sur=r_high, m_l = np.logspace(0,6,100))
tinker = mf.Tinker(sur=r_high, m_l = np.logspace(0,6,100))
wdm_s = mf.WDM_stream(sur=r_high, m_l = np.logspace(0,6,100))
wdm_l = mf.WDM_lensing(sur=r_high, m_l = np.logspace(0,6,100))
power_law = mf.PowerLaw(sur=r_high, m_l = np.logspace(0,6,100))
pbh = mf.PBH(sur=r_high, m_l = np.logspace(0,6,100))
press_s = mf.PressSchechter_test(sur=r_high, m_l = np.logspace(0,6,100))
```

```python
%%time
cdm = mf.CDM_Test(sur=r_norm, m_l = np.logspace(0,6,100))
tinker = mf.Tinker(sur=r_norm, m_l = np.logspace(0,6,100))
wdm_s = mf.WDM_stream(sur=r_norm, m_l = np.logspace(0,6,100))
wdm_l = mf.WDM_lensing(sur=r_norm, m_l = np.logspace(0,6,100))
power_law = mf.PowerLaw(sur=r_norm, m_l = np.logspace(0,6,100))
pbh = mf.PBH(sur=r_norm, m_l = np.logspace(0,6,100))
press_s = mf.PressSchechter_test(sur=r_norm, m_l = np.logspace(0,6,100))
```

```python
plt.scatter(cdm.m_l,cdm.n_l,label='CDM')
plt.scatter(tinker.m_l,tinker.n_l,label='Tinker')
plt.scatter(press_s.m_l,press_s.n_l,label='Press S')
plt.scatter(wdm_s.m_l,wdm_s.n_l,label='WDM S')
plt.scatter(wdm_l.m_l,wdm_l.n_l,label='WDM L')
plt.scatter(pbh.m_l,pbh.n_l,label='PBH')
plt.scatter(power_law.m_l,power_law.n_l,label='Power Law')
plt.title('Predicted number of lenses Higher Volume')
plt.xlabel(r'Subhalo mass $M_{sun}$')
plt.ylabel('Number of lenses')
plt.xscale('log')
plt.legend()
plt.show()
```
